# Importing

```python
import IPython
from scipy.io.wavfile import read, write
import noisereduce as nr
import soundfile as sf
import matplotlib.pyplot as plt
import numpy as np
import io
%matplotlib inline
```

# Loading Data

```python
with open("1.wav", "rb") as wavfile:
    input_wav = wavfile.read()

data, rate = sf.read(io.BytesIO(input_wav))

IPython.display.Audio(data=data, rate=rate)
```

# Visual Representation

```python
fig, ax = plt.subplots(figsize=(20,3))
ax.plot(data)
```

# Extracting Noise

```python
noisy_part = data[35000:]
IPython.display.Audio(data=noisy_part, rate=rate)
```

# Reducing Noise

```python
noise_reduced = nr.reduce_noise(audio_clip=data, noise_clip=noisy_part, prop_decrease=1.0, verbose=True)
```

# Audio Visual Inspection

```python
IPython.display.Audio(data=noise_reduced, rate=rate)
```

```python
fig, ax = plt.subplots(figsize=(20,3))
ax.plot(noise_reduced)
```

# Saving to New file

```python
write("1_denoised.wav", rate, noise_reduced)
```
